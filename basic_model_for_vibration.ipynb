The simple model of a vibrating mechanical system has the following form:

d²u/dt + w² u = 0

u(0) = I, du(0)/dt = 0

t ∈ (0, T]

The exact solution of equation is:

u(t) = I cos(w t)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
T = 5.0
I = 1.0
w = 5.0
Nt = 10001
t = np.linspace(0, T, Nt + 1)
dt = t[1] - t[0]
```

```python
u = np.zeros_like(t)

u[0] = I
u[1] = u[0] - 0.5 * dt**2 * w**2 * u[0]

for n in range(1, Nt):
  u[n + 1] = 2 * u[n] - u[n - 1] - dt**2 * w**2 * u[n]
```

```python
plt.plot(t, u)
plt.show()
```
